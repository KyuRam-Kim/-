# <h1> 1-3 마켓과 머신러닝 </h1>
<h3> 생선 분류 문제 </h3>


*   맷플롯립 (Matplotlib) 패키지 사용 (matplotlib 패키지 하위 pyplot 모듈 사용)
*   사이킷런 (Scikit-Learn) 패키지 사용 (sklearn.neighbors 모듈 하위 KNeighborsClassifier 클래스 사용)

K-최근접 이웃 모델 (K-Nearest Neighbors)


*   알고리즘 훈련이라고 해봤자 데이터의 저장에 불과함
*   저장된 데이터를 기반으로 새로운 데이터를 비교하는 것 뿐
*   새로운 데이터에 대해 예측할때 가장 가까운 직선거리에 어떤 데이터가 있는지 살핌
*   단점

  1.   데이터가 아주 많은 경우 사용하기 어려움
  2.   데이터가 크기 때문에 메모리가 많이 필요하고 직선거리를 계산하는데도 많은 시간 필요

```python
#도미 길이,무게 산점도

bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]   #도미의 길이
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]     #도미의 무게

import matplotlib.pyplot as plt   #matplotlib의 pyplot 함수를 plt로 줄여서 사용함   
                                  # from matplotlib import pyplot as plt 도 가능

plt.scatter(bream_length,bream_weight)
plt.xlabel('length')  #X축은 길이
plt.ylabel('weight')  #Y축은 무게
plt.show()

#도미와 빙어 길이, 무게 산점도

smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]  #빙어 길이
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]         #빙어 무게

plt.scatter(bream_length,bream_weight)
plt.scatter(smelt_length,smelt_weight)
plt.xlabel('length')  #X축은 길이
plt.ylabel('weight')  #Y축은 무게
plt.show()

#사이킷런 패키지에서 K-최근접 이웃 알고리즘 구현하기

length = bream_length + smelt_length    #길이 데이터 합치기
weight = bream_weight + smelt_weight    #무게 데이터 합치기
#print(length)  
#print(weight)  #합친 데이터들 확인

fish_data = []
for l,w in zip(length,weight):                        #for 반복문: 동일한 작업 계속 반복
  fish_data.append([l,w])                             #zip() 함수: 나열된 리스트에서 원소를 하나씩 꺼내줌
print(fish_data)

fish_target = [1]*35 + [0]*14                #정답지 만들기, 도미(찾으려는 대상)을 1로 지정, 빙어를 0으로 지정
print(fish_target)

from sklearn.neighbors import KNeighborsClassifier    #KNeighborsClassifier 임포트
kn=KNeighborsClassifier()                             #KNeighborsClassifier 객체 만들기

kn.fit(fish_data,fish_target)                         #fit() method로 알고리즘 훈련 (훈련이라고 해봤자 데이터를 저장하는게 전부임)
kn.score(fish_data,fish_target)                       #정확도, score() method는 0에서 1사이의 값을 반환함

print(kn.predict([[30,600]]))                         #길이 30cm, 무게 600g의 생선 예측하기   #array: 배열

print(kn._fit_X)
print(kn._y)

kn49 = KNeighborsClassifier(n_neighbors=49)           #참고 데이터를 49개로 한 kn49 모델 (매개변수 기본값은 5개)

kn49.fit(fish_data,fish_target)                       #kn49 훈련  (이 알고리즘의 훈련은 데이터를 저장하는게 전부임)
kn49.score(fish_data,fish_target)                     #kn49의 정확도 반환


```

# <h1> 2-1 훈련세트와 테스트 세트 </h1>

훈련세트/테스트세트에 샘플이 골고루 섞여있어야 함





```python
fish_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0, 9.8, 
                10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
fish_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0, 6.7, 
                7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

fish_data = [[l,w] for l,w in zip(fish_length,fish_weight)]     #49개의 샘플 생성, 샘플의 특성은 2개 (길이,무게)
fish_target = [1]*35 + [0]*14

#49개의 샘플 중 처음 35개를 훈련세트로, 나머지 14개를 테스트 세트로 사용하자

from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()

train_input = fish_data[0:35]         #훈련세트로 입력값 중 0번부터 34번 인덱스까지 사용 (35개)
train_target = fish_target[0:35]      #훈련세트로 타깃값 중 0번부터 34번 인덱스까지 사용 (35개)

test_input = fish_data[35:]           #테스트 세트로 입력값 중 35번부터 마지막 인덱스까지 사용 (14개)
test_target = fish_target[35:]        #테스트 세트로 입력값 중 35번부터 마지막 인덱스까지 사용 (14개)

#################여기서 잠깐 인덱스란?
#인덱스: 배열의 위치

print(fish_data[4])     #5번 인덱스 샘플 (인덱스는 0부터 시작)

print(fish_data[0:5])   #0번 인덱스 샘플부터 4번 인덱스 샘플까지 출력 (마지막 인덱스의 원소는 포함되지 않음)      #여기서 나온 콜론(:)은 슬라이싱 연산자, 범위를 지정함
print(fish_data[:5])    #첫 샘플부터 하는거면 이런식으로 생략 가능

print(fish_data[44:])   #44번 인덱스 샘플부터 마지막 샘플까지 출력
#################다시 본론으로...

kn = kn.fit(train_input, train_target)    #훈련 세트로 훈련
kn.score(test_input, test_target)        #테스트 세트로 평가

#정확도 0
#테스트 세트에 빙어만 들어가 있음
#샘플링 편향을 해결해보자


#넘파이
import numpy as np        #넘파이 임포트

input_arr = np.array(fish_data)       #파이썬 리스트를 넘파이 배열로 바꾸기
target_arr = np.array(fish_target)     #넘파이 array() 함수에 전달하면 된다!

print(input_arr)                      #입력값 넘파이 배열 출력해보기
print(input_arr.shape)                #입력값 넘파이 배열의 shape 속성 출력하기
                                      #(샘플 수, 특성 수)를 출력해줌 (행 갯수, 열 갯수)

np.random.seed(42)                    #교재와 같은 결과를 출력하기 위해 랜덤시드를 42로 지정
index = np.arange(49)                 #0부터 48까지 1씩 정수들이 증가하는 배열 생성
np.random.shuffle(index)              #해당 배열 셔플
print(index)                          #섞인 배열 출력

#넘파이에서 배열 인덱싱 기능 사용
#여러개의 인덱스로 한번에 여러 개의 원소 선택 가능

#Example
print(input_arr[[1,3]])               #1번 인덱스, 3번 인덱스 배열 출력
#

train_input = input_arr[index[:35]]   #35개 샘플 훈련 세트로 만들기
train_target = target_arr[index[:35]]

test_input = input_arr[index[35:]]    #나머지 14개 샘플을 테스트 세트로 만들기
test_target = target_arr[index[35:]]


#훈련세트, 테스트 세트에 도미와 빙어가 잘 섞여있는지 산점도로 그려보자

import matplotlib.pyplot as plt
plt.scatter(train_input[:,0], train_input[:,1])   #train_input[1:49 , 0] 과 같음
                                                  #슬라이싱 연산자 : 로 모든 원소 선택할 경우 시작/종료 인덱스 모두 생략 가능
                                                  #1차원(2차원 배열이라 여기서 표기 안함), 모든 행들, 0열/1열 
plt.scatter(test_input[:,0],test_input[:,1])
plt.xlabel("Length")
plt.ylabel("Weight")
plt.show()                                


##### 2번째 머신러닝 프로그램 #####
#새롭게 만든 훈련세트, 테스트세트로 새로운 모델을 훈련시키자

kn = kn.fit(train_input,train_target) 
kn.score(test_input, test_target)               #정확도 1 출력

print(kn.predict(test_input))                   #테스트 세트의 예측 결과 확인
print(test_target)                              #실제 타깃 확인

```

# <h1> 2-2 데이터 전처리(Data Preprocessing) </h1>

-> 머신러닝 모델에 훈련 데이터를 주입하기 전에 가공하는 단계

-> 샘플 간의 거리에 영향을 많이 받는 알고리즘을 제대로 사용하려면 특성값을 일정한 기준으로 맞춰주어야함


---


<h3> 생선 분류 문제 </h3>


*  넘파이 (Numpy) 패키지 사용
  
  -브로드캐스팅 : 넘파이 배열에서 자동으로 사칙연산을 모든 행이나 열로 확장하는 기능
*  사이킷런 (Scikit-Learn) 패키지 사용 (sklearn.model_selection 모듈 하위 train_test_split() 함수 사용)


---

1. 데이터가 클수록 파이썬 리스트는 비효율적. 저수준 언어로 개발되어 빠르고 데이터 과학 분야에 최적화된 넘파이 배열을 사용하는 것이 좋음

2. 스케일이 다른 특성을 일정하게 맞춰줘야함
  -> 표준점수로 변환

3. **데이터를 전처리할때 훈련세트 뿐만 아니라 테스트 세트도 표준화해야하는 것 주의!!!**



```python
#넘파이로 데이터 준비하기
fish_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0, 9.8, 
                10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
fish_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0, 6.7, 
                7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

import numpy as np

np.column_stack(([1,2,3],[4,5,6]))    #numpy의 column_stack 함수
                                      #전달받은 리스트를 일렬로 세운 다음 차례대로 나란히 연결한다
                                      #연결할 리스트는 파이썬 튜플로 전달함

fish_data = np.column_stack((fish_length, fish_weight))
print(fish_data)

#원하는 갯수의 1 또는 0을 채우는 ones/zeros 함수
#첫번째 차원을 따라 배열을 연결하는 concatenate 함수
fish_target = np.concatenate((np.ones(35), np.zeros(14)))
print(fish_target)

#############################################################################

#사이킷런으로 훈련세트와 테스트세트 나누기
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    fish_data, fish_target, random_state = 42)        #random_state 매개변수로 랜덤시드 지정(교재와 동일하게 하기 위해)
                                                      #이 함수는 기본적으로 25%를 테스트 세트로 떼어냄

print(train_input.shape, test_input.shape)            #2차원 배열의 입력 데이터
print(train_target.shape, test_target.shape)          #1차원 배열의 타깃 데이터                 

print(test_target)                #도미와 빙어의 비율이 3.3 : 1 로 나타남 -> 샘플링 편향

train_input, test_input, train_target, test_target = train_test_split(
    fish_data, fish_target, stratify=fish_target, random_state=42)
print(test_target)                #stratify 매개변수에 타깃 데이터를 전달하면 클래스 비율에 맞게 데이터를 나눔

#############################################################################

#수상한 도미 한마리
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()

kn.fit(train_input,train_target)
kn.score(test_input, test_target)

print(kn.predict([[25,150]]))         #[0.] 출력, 빙어로 예측

import matplotlib.pyplot as plt       #문제 해결을 위해 산점도를 그려보자
plt.scatter(train_input[:,0], train_input[:,1])     
plt.scatter(25,150,marker="^")        #marker 매개변수를 "^"로 지정하여 삼각형으로 표시
plt.xlabel("Length")
plt.ylabel("Weight")
plt.show()

distances, indexes = kn.kneighbors([[25,150]])    #kneighbors() 메서드 : 이웃까지의 거리와 이웃들의 인덱스 반환

plt.scatter(train_input[:,0], train_input[:,1])     
plt.scatter(25,150,marker="^")      
plt.scatter(train_input[indexes,0], 
            train_input[indexes,1], marker = "D")    #marker 매개변수를 "D"로 지정하여 다이아몬드로 표시
plt.xlabel("Length")
plt.ylabel("Weight")
plt.show()
#5개의 샘플 중 가장 가까운 도미를 제외하고 나머지 4개는 빙어

print(train_input[indexes])
print(train_target[indexes])
print(distances)

#############################################################################

#기준을 맞춰라
#x축, y축 범위가 다르기 때문에 범위를 일정하게 지정하여 산점도를 그려보자
plt.scatter(train_input[:,0], train_input[:,1])     
plt.scatter(25,150,marker="^")      
plt.scatter(train_input[indexes,0], 
            train_input[indexes,1], marker = "D")
plt.xlim((0,1000))    
plt.xlabel("Length")
plt.ylabel("Weight")
plt.show()
#산점도가 거의 일직선상으로 표현됨 -> 생선의 길이는 영향을 끼치지 못하고 오로지 무게(y축)만 고려 대상이 됨
#두 특성의 값이 놓인 범위가 다르다 : 두 특성의 스케일(scale)이 다르다
#넘파이를 이용하여 표준점수를 구한다

mean = np.mean(train_input, axis=0)   #평균을 계산하는 np.mean() 함수
std = np.std(train_input, axis=0)     #표준편차를 계산하는 np.std() 함수
#각 특성마다 값의 스케일이 다르므로 각 특성별로 통계량을 구해야함
#매개변수 axis=0 으로 지정하면 행을따라 각 열의 통계 값을 계산함

print(mean,std)               #계산한 평균과 표준편차 확인

train_scaled = (train_input-mean)/std     #표준화
#브로드캐스팅이라는 넘파이 기능이 적용됨. 각 행마다 표준화 작업 진행

#############################################################################

#전처리 데이터로 모델 훈련하기
#표준화된 데이터를 산점도에 나타내어 보자

plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(25,150,marker="^")              #우리가 보려는 생선의 데이터도 표준화 필요
plt.xlabel("Length")
plt.ylabel("Weight")
plt.show()

new = ([25,150]-mean)/std                   #훈련세트의 평균/표준편차를 이용하여 표준화
plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(new[0],new[1],marker="^")              
plt.xlabel("Length")
plt.ylabel("Weight")
plt.show()

#이제 이 데이터셋으로 k-최근접 이웃 모델을 훈련시켜보자
kn.fit(train_scaled, train_target)

#테스트 세트로 평가할때 주의 : 테스트 세트도 표준화된 데이터여야 함
test_scaled = (test_input-mean)/std     #훈련세트의 평균/표준편차 이용
print(kn.score(test_scaled, test_target))     #테스트 세트로 모델 평가

print(kn.predict([new]))                 # [1.] 출력, 도미로 예측

#kneighbors() 함수로 이 샘플의 최근접 이웃을 구하고 산점도로 그려보자
distances, indexes = kn.kneighbors([new])
plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(new[0],new[1],marker="^")
plt.scatter(train_scaled[indexes,0], train_scaled[indexes,1], marker="D")             
plt.xlabel("Length")
plt.ylabel("Weight")
plt.show()
#5개의 최근접 이웃 모두 도미

#############################################################################
```
